## 1. Data Preparation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn
import seaborn as sns
```

```python
review_data = pd.read_csv('Reviews.csv')
review_data.head()
```

```python
review_data.shape
```

```python
review_data.info()
```

```python
review_data.isnull().sum()
```

### While looking at the number of null values in the dataframe, it is interesting to note that $\sim$0% of the entries are not assigned Summary to particular product. With the data available, it is impossible to impute values for these entries are thus useless for the current exercise. So I delete them from the dataframe:

```python
review_data.dropna(axis = 0, subset = ['Summary'], inplace = True)
```

```python
review_data.shape
```

```python
review_data.dropna(axis = 0, subset = ['ProfileName'], inplace = True)
```

```python
review_data.shape
```

```python
review_data.head()
```

```python
review_data.tail()
```

### OK, therefore, by removing these entries we end up with a dataframe filled at 100% for all variables! Finally, I check for duplicate entries and delete them:

```python
print('Entries duplicate: {}'.format(review_data.duplicated().sum()))
review_data.drop_duplicates(inplace = True)
```

### No duplicate Entries

## 2. Exploring the content of Dataset

This dataset consists of reviews of fine foods from amazon. The data span a period of more than 10 years, including all ~500,000 reviews up to October 2012. Reviews include product and user information, ratings, and a plain text review. It also includes reviews from all other Amazon categories.

**Data includes:**

* Reviews from Oct 1999 - Oct 2012
* 568,454 reviews
* 256,059 users
* 74,258 products
* 260 users with > 50 reviews

## Columns :
* ProductId - Unique identifier for the product
* UserId - Unique identifier for the user
* ProfileName - Profile name of User
* HelpfulnessNumerator - Num of users who found review helpful
* HelpfulnessDenomitaor - Number of users who indicated whether they found the review helpful or not
* Score - Rating between 1 to 5
* Time - Timestamp for the review
* Summary - Brief summary of the review
* Text - Text of the review

## 3. Data Visualization

```python
sns.distplot(review_data['Time'],bins = 50)
```

## Highest reviews by ProfileName. 

```python
profilename = review_data['ProfileName'].value_counts().head(10)
profilename
```

### Highest reviews are given by C. F. Hill "CFH" which is about 451 reviews

```python
sns.barplot(y=profilename.index,x=profilename,palette='Blues_d')
```

## Highest review contribution by ProfileName 

```python
profile = review_data.groupby('ProfileName').filter(lambda x : len(x) > 300 )
len(profile)
```

```python
profilegroup = profile.groupby('ProfileName')['Score'].count()
profilegroup
```

```python
plt.figure(figsize = (20,5))
sns.countplot(x='ProfileName',hue = 'Score',data = profile,palette='Blues_d')
```

# Rating Score Count.
